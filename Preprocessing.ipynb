```python
import pandas as pd
import re
import nltk
nltk.download('stopwords')

def _remove_digits(text):
    return re.sub("[0-9]", "", text)

def _remove_english_chars(text):
    return re.sub("[a-zA-Z]", "", text)

def _remove_links(text):
    text = re.sub(r"http\S+", " ", text, flags=re.MULTILINE)
    return text

def _remove_twitter_meta(text):
    text = re.sub("(@[A-Za-z0-9]+)", " ", text)
    text = re.sub("(#[A-Za-z0-9]+)", " ", text)
    text = _remove_links(text)
    return text

def _remove_special_chars(text, excluded_chars=[]):
    regex_special_chars = "\\^+*[]-)(,:;''و في الى"
    ignored_chars = ""
    for char in excluded_chars:
        if char in regex_special_chars:
            ignored_chars += f"\\" + char
        else:
            ignored_chars += char
    return re.compile("([^\n\u0621-\u064A0-9a-zA-Z " + ignored_chars + "])").sub(" ", text)

def _remove_extra_spaces(text):
    text = re.sub(" +", " ", text)
    return text

def remove_stopwords_arabic(text):
    # Download the stopwords corpus if not already downloaded
    nltk.download('stopwords')

    # Load the Arabic stopwords
    stopwords = nltk.corpus.stopwords.words('arabic')

    # Tokenize the text into individual words
    words = nltk.word_tokenize(text)

    # Remove stopwords from the text
    filtered_words = [word for word in words if word.lower() not in stopwords]

    # Join the filtered words back into a single string
    filtered_text = ' '.join(filtered_words)

    return filtered_text



def count_words(text):
    words = re.findall(r'\w+', text)
    return len(words)



# Load the NfoodBlogs DataFrame
NfoodBlogs = pd.read_csv('food_blogs.csv')

# Apply cleaning functions to each row in the 'Blog' column
cleaned_blogs = []
word_counts = []
unique_word_counts = []

for index, row in NfoodBlogs.iterrows():
    text = row['Blog']
    text = _remove_digits(text)
    text = _remove_english_chars(text)
    text = _remove_twitter_meta(text)
    text = _remove_links(text)
    text = _remove_special_chars(text)
    text = _remove_extra_spaces(text)
    text = remove_stopwords_arabic(text)
 
    cleaned_blogs.append(text)
    word_counts.append(count_words(text))


# Create a new DataFrame with cleaned blogs and word counts
cleaned_df = pd.DataFrame({'Cleaned_Blog': cleaned_blogs, 'Word_Count': word_counts})

# Save the cleaned blogs and word counts to a CSV file
cleaned_df.to_csv('cleaned_blogs.csv', index=False, encoding='utf-8-sig')
```

# TF-IDF

```python
clean['TF'] = clean['Tokens'].apply(lambda token: {word: token.count(word)/len(token) for word in token})
```

```python
clean
```

```python
# Count document frequency (DF) for each unique word
document_frequency = {}
for tokens in clean['Tokens']:
    unique_tokens = set(tokens)
    for token in unique_tokens:
        document_frequency[token] = document_frequency.get(token, 0) + 1
```

```python
import math
# Calculate inverse document frequency (IDF)
N = len(clean)# 1714
clean['IDF'] = clean['Tokens'].apply(lambda x: {word: math.log(N / (1 + document_frequency[word])) for word in x})
```

```python
# Compute TF-IDF
clean['TF-IDF'] = clean.apply(lambda row: {word: row['TF'].get(word, 0) * row['IDF'].get(word, 0) for word in row['Tokens']}, axis=1)
```

```python
def get_vectors(tokens):
    text = ' '.join(tokens)
    doc = nlp(text)
    token_vector_dict = {}
    for token in doc:
        token_vector_dict[token.text] = token.vector
    return token_vector_dict

# Apply the function to the 'Tokens' column
clean['Token_Vectors'] = clean['Tokens'].apply(get_vectors)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
# Load the BERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
model = AutoModel.from_pretrained("bert-base-multilingual-cased")
```

```python
def apply_bert_embedding(sentence):
    # Tokenize the sentence
    tokens = tokenizer.tokenize(sentence)

    # Convert tokens to token IDs
    token_ids = tokenizer.convert_tokens_to_ids(tokens)

    # Add special tokens and create input tensors
    input_ids = [tokenizer.cls_token_id] + token_ids + [tokenizer.sep_token_id]
    input_tensor = torch.tensor([input_ids])

    # Forward pass through BERT model to get embeddings
    outputs = model(input_tensor)
    embeddings = outputs.last_hidden_state

    # Remove special tokens from embeddings
    token_embeddings = embeddings[0][1:-1]

    # Convert embeddings to NumPy array
    token_vectors = token_embeddings.detach().numpy()

    return token_vectors
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


x1 = apply_bert_embedding('الشيف')
x2 = apply_bert_embedding('الشيف')

# Calculate the cosine similarity between x1 and x2
similarity = cosine_similarity(x1, x2)

similarity[0][0]
```
