```python
import requests
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from urllib.parse import urljoin, quote

from fake_useragent import UserAgent
import pandas as pd
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
```

```python
def generate_user_agent():
    user_agent = UserAgent()
    return user_agent.random

def extract_urls(driver, num_of_articles=40):
    urls = [i.get_attribute("href") for i in WebDriverWait(driver, 5).until(EC.presence_of_all_elements_located((By.XPATH, "//a[contains(@href, '/article/')]")))]
    while len(urls) < num_of_articles:
        try:
            driver.execute_script("document.getElementById('article_loadmore').click();")
            urls = [i.get_attribute("href") for i in WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, "//a[contains(@href, '/article/')]/..")))]
        except Exception as e:
            print(e)
            break
    return [i.get_attribute("href") for i in driver.find_elements(By.XPATH, "//a[contains(@href, '/article/')]")]

def extract_paragraphs(url):
    headers = {'User-Agent': generate_user_agent()}
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    soup = BeautifulSoup(response.content, 'lxml')
    paragraphs = soup.find_all('div', class_='article_content')
    return paragraphs

s=Service('C://Users//HP//Documents//documents project//chromedriver.exe')
driver=webdriver.Chrome(service=s)
driver.get("https://www.atyabtabkha.com/tag/%d9%85%d8%af%d9%88%d9%86%d8%a9")
urls = extract_urls(driver, num_of_articles=7000)
driver.quit()
all_blogs = []

# Iterate over the URLs, extract and store the paragraphs
for url in urls:
    paragraphs = extract_paragraphs(url)
    print(f"Completed {urls.index(url)}/{len(urls)}")
    for paragraph in paragraphs:
        blog = paragraph.get_text().strip('                 ')
        all_blogs.append(blog)
        
# Create a dictionary to store the blogs
data = {"blogs": all_blogs}
with open('food_blogs.csv', 'w', newline='', encoding='utf-8-sig') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Blog'])
    writer.writerows([[blog] for blog in all_blogs])
```
